```python
import pandas as pd 
import numpy as np
```

```python
# import csv file
rates = pd.read_csv("/Users/victoriaguo/Desktop/DS 4002/project2/10year_rates.csv")
```

```python
rates.head()
```

```python
# only keep rows that correspond to Treasury Bills, Bonds, and Notes
conditions = (rates['Security Description'] == 'Treasury Bills') | (rates['Security Description'] == 'Treasury Notes') | (rates['Security Description'] == 'Treasury Bonds')

rates_real = rates[conditions]
```

```python
rates_real.head()
```

```python
# keep necessary columns: security description, record date, and average interest rate amount
columns_keep = ['Security Description', 'Record Date', 'Average Interest Rate Amount']

# remove na's 
rates_final = rates_real[columns_keep].dropna()
```

```python
rates_final.info()
rates_final.head()
```

```python
# export data to csv file

rates_final.to_csv("/Users/victoriaguo/Desktop/DS 4002/project2/final_data.csv")
```
